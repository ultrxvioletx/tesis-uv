```python
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 15, 1000)
cmap = plt.cm.Accent

with h5py.File('dset.h5', 'r') as f:
    def structure(name, obj):
        print(name)
        if isinstance(obj, h5py.Dataset):
            print(f"    Shape: {obj.shape}, Dtype: {obj.dtype}")
    f.visititems(structure)
    
    dataset = f['dset']
    data = dataset[:]
P_gg = data[:,0]
P_ge = data[:,1]
P_eg = data[:,2]
P_ee = data[:,3]
N_expect = data[:,4]
X_expect = data[:,5]
P_expect = data[:,6]
```

```python
# probabilidades
for i in range(4):
    plt.plot(t, data[:,i], label=f'|{i}>', color=cmap(i), linewidth=2)
plt.xlabel('tiempo [$1/Ω_0$]')
plt.ylabel('probabilidad (estado del átomo)')
plt.grid(True, color="gray", alpha=0.4)
plt.legend()
plt.show()
```

```python
# número esperado de fotones
plt.plot(t, N_expect, label='<N>', color=cmap(0))
plt.xlabel('tiempo [$1/Ω_0$]')
plt.ylabel('promedio de fotones (cavidad)')
plt.legend()
plt.grid()
plt.show()
```

```python
# número esperado de la posición
plt.plot(t, X_expect, label='<X>', color=cmap(0))
plt.xlabel('tiempo [$1/Ω_0$]', fontsize=12)
plt.ylabel('valor esperado posición', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
# número esperado del momento
plt.plot(t, P_expect, label='<P>', color=cmap(0))
plt.xlabel('tiempo [$1/Ω_0$]', fontsize=12)
plt.ylabel('valor esperado momento', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```
